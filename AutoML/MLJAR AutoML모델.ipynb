## Compete모드, explain모드, perform모드 차이점


### 매개변수 설명
- `results_path`: 결과를 저장할 경로입니다. 기본값은 `None`으로, 이 경우 현재 작업 디렉토리에 결과가 저장됩니다.
- `total_time_limit`: AutoML이 실행되는 총 시간 제한(초)입니다. 기본값은 3600초(1시간)입니다.
- `mode`: AutoML의 작동 모드입니다. 'Explain', 'Perform', 'Compete', 'Optuna' 중 하나를 선택할 수 있습니다. 기본값은 'Explain'입니다.
- `ml_task`: 머신러닝 작업의 종류입니다. 'auto', 'binary_classification', 'multiclass_classification', 'regression' 중 하나를 선택할 수 있습니다. 기본값은 'auto'로, 이 경우 데이터를 분석하여 작업 종류를 자동으로 결정합니다.
- `model_time_limit`: 각 모델의 학습 시간 제한(초)입니다. 기본값은 `None`으로, 이 경우 각 모델의 학습 시간 제한이 없습니다.
- `algorithms`: 사용할 알고리즘 목록입니다. 'auto' 또는 ['Baseline', 'Linear', 'Decicion Tree', 'Random Forest', 'Extra Trees', 'LightGBM', 'Xgboost', 'CatBoost', 'Neural Network', 'Nearest Neighbors'] 중 하나 이상을 선택할 수 있습니다. 기본값은 'auto'로, 이 경우 모든 알고리즘이 사용됩니다.
- `train_ensemble`: 앙상블 모델을 학습할지 여부입니다. 기본값은 `True`로, 이 경우 앙상블 모델이 학습됩니다.
- `stack_models`: 스택 모델을 사용할지 여부입니다. 'auto' 또는 `True`/`False`를 선택할 수 있습니다. 기본값은 'auto'로, 이 경우 Compete 모드에서만 스택 모델이 사용됩니다.
- `eval_metric`: 평가 지표입니다. 지원되는 지표는 작업 종류에 따라 다릅니다. 기본값은 'auto'로, 이 경우 작업 종류에 따라 적절한 지표가 자동으로 선택됩니다.
- `validation_strategy`: 검증 전략입니다. 'auto' 또는 검증 전략에 대한 사전을 선택할 수 있습니다. 기본값은 'auto'로, 이 경우 5-겹 교차 검증이 사용됩니다.
- `explain_level`: 설명 수준입니다. 0, 1, 2 중 하나를 선택할 수 있습니다. 기본값은 'auto'로, 이 경우 Explain 모드에서는 2가, 그 외의 모드에서는 1이 사용됩니다.
- `golden_features`: 골든 피처 생성 여부 또는 생성할 골든 피처의 수입니다. 'auto', `True`/`False`, 또는 정수를 선택할 수 있습니다. 기본값은 'auto'로, 이 경우 Compete 모드에서만 골든 피처가 생성됩니다.
- `features_selection`: 피처 선택 여부입니다. 'auto' 또는 `True`/`False`를 선택할 수 있습니다. 기본값은 'auto'로, 이 경우 Compete 모드에서만 피처 선택이 수행됩니다.
- `start_random_models`: 초기에 학습할 무작위 모델의 수입니다. 기본값은 'auto'로, 이 경우 작업 종류와 데이터 크기에 따라 적절한 값이 자동으로 결정됩니다.
- `hill_climbing_steps`: 힐 클라이밍 단계 수입니다. 기본값은 'auto'로, 이 경우 Compete 모드에서만 힐 클라이밍이 수행됩니다.
- `top_models_to_improve`: 개선할 상위 모델의 수입니다. 기본값은 'auto'로, 이 경우 Compete 모드에서만 상위 모델이 개선됩니다.
- `boost_on_errors`: 오류에 대한 부스트 여부입니다. 'auto' 또는 `True`/`False`를 선택할 수 있습니다. 기본값은 'auto'로, 이 경우 Compete 모드에서만 오류에 대한 부스트가 수행됩니다.
- `kmeans_features`: K-Means 클러스터링 알고리즘을 사용하여 새로운 피처를 생성할지 여부입니다. 'auto' 또는 `True`/`False`를 선택할 수 있습니다. 기본값은 'auto'로, 이 경우 Compete 모드에서만 K-Means 피처가 생성됩니다.
- `mix_encoding`: 범주형 피처에 대해 다양한 인코딩 방법을 혼합하여 사용할지 여부입니다. 'auto' 또는 `True`/`False`를 선택할 수 있습니다. 기본값은 'auto'로, 이 경우 Compete 모드에서만 다양한 인코딩 방법이 혼합되어 사용됩니다.
- `max_single_prediction_time`: 단일 예측의 최대 시간(초)입니다. 기본값은 `None`으로, 이 경우 단일 예측의 최대 시간 제한이 없습니다.
- `optuna_time_budget`: Optuna 하이퍼파라미터 최적화의 시간 제한(초)입니다. 기본값은 `None`으로, 이 경우 Optuna 하이퍼파라미터 최적화의 시간 제한이 없습니다.
- `optuna_init_params`: Optuna 하이퍼파라미터 최적화의 초기 매개변수입니다. 기본값은 빈 사전입니다.
- `optuna_verbose`: Optuna 하이퍼파라미터 최적화의 출력 여부입니다. 기본값은 `True`로, 이 경우 Optuna 하이퍼파라미터 최적화의 출력이 표시됩니다.
- `n_jobs`: 병렬 작업의 수입니다. 기본값은 -1로, 이 경우 사용 가능한 모든 프로세서가 사용됩니다.
- `verbose`: 출력 수준입니다. 0, 1, 2 중 하나를 선택할 수 있습니다. 기본값은 1로, 이 경우 일반적인 출력이 표시됩니다.
- `random_state`: 난수 생성기의 초기 상태입니다. 기본값은 1234입니다.

- `results_path`: 결과를 저장할 경로입니다. 기본값은 `None`으로, 이 경우 현재 작업 디렉토리에 `AutoML_{number}` 형식의 이름으로 디렉토리가 생성됩니다. 여기서 `number`는 1에서 1000 사이의 값으로, 사용 가능한 디렉토리 이름에 따라 결정됩니다. 만약 `results_path`가 AutoML 결과가 있는 디렉토리(`params.json` 파일이 존재해야 함)를 가리키면, 모든 모델이 로드됩니다.
- `total_time_limit`: AutoML 학습의 총 시간 제한(초)입니다. `model_time_limit`이 `None`이 아닌 경우 사용되지 않습니다.
- `mode`: AutoML의 목표와 검색 강도를 결정하는 매개변수입니다. `Explain`, `Perform`, `Compete`, `Optuna` 중 하나를 선택할 수 있습니다.
    - `Compete`: 머신러닝 경쟁(최대 성능)을 위해 사용됩니다.
        - 80/20 학습/테스트 분할 또는 5-겹 CV 또는 10-겹 CV를 사용합니다. 이는 `total_time_limit`에 따라 다릅니다. 직접 설정하지 않으면 AutoML이 검증 방법을 자동으로 선택합니다.
        - 다음 모델을 사용합니다: `Decision Tree`, `Random Forest`, `Extra Trees`, `LightGBM`, `XGBoost`, `CatBoost`, `Neural Network`, `Nearest Neighbors`, `Ensemble` 및 `Stacking`.
        - 보고서에는 학습 곡선만 포함됩니다.
    - `Optuna`: Optuna 하이퍼파라미터 최적화를 사용하여 최적의 모델을 찾습니다.
        - 10-겹 교차 검증(CV)을 사용합니다.
        - 다음 알고리즘을 Optuna로 조정합니다: `Random Forest`, `Extra Trees`, `LightGBM`, `XGBoost`, `CatBoost`, `Neural Network`.
        - 학습된 모델에 대해 `Ensemble` 및 `Stacking`을 적용합니다.
        - 보고서에는 학습 곡선만 포함됩니다.

- `ml_task`: 머신러닝 작업의 종류입니다. 'auto', 'binary_classification', 'multiclass_classification', 'regression' 중 하나를 선택할 수 있습니다.
    - 'auto'로 설정하면 AutoML이 목표 값에 따라 작업 종류를 추측하려고 합니다.
    - 목표 값이 2개만 있는 경우 작업 종류가 `"binary_classification"`으로 설정됩니다.
    - 목표 값의 개수가 2개에서 20개(포함) 사이인 경우 작업 종류가 `"multiclass_classification"`으로 설정됩니다.
    - 그 외의 경우 작업 종류가 `"regression"`으로 설정됩니다.

- `model_time_limit`: 단일 모델 학습의 시간 제한(초)입니다. `model_time_limit`이 설정되면 `total_time_limit`은 적용되지 않습니다. 단일 모델은 여러 학습기를 포함할 수 있습니다. 후속 학습기의 시간 제한은 `model_time_limit`을 기반으로 계산됩니다.
    - 예를 들어, 10-겹 교차 검증의 경우 하나의 모델에 10개의 학습기가 있습니다. `model_time_limit`은 10개의 학습기 모두에 대한 시간입니다.

- `algorithms`: 학습에 사용할 알고리즘 목록입니다. 알고리즘은 다음과 같습니다:
    - `Baseline`
    - `Linear`
    - `Decision Tree`
    - `Random Forest`
    - `Extra Trees`
    - `LightGBM`
    - `Xgboost`
    - `CatBoost`
    - `Neural Network`
    - `Nearest Neighbors`
- `train_ensemble`: 학습이 끝난 후 앙상블이 생성되는지 여부입니다. 기본값은 `True`로, 이 경우 앙상블이 생성됩니다.
- `stack_models`: 학습이 끝난 후 모델 스택이 생성되는지 여부입니다. 스택 레벨은 1입니다. 기본값은 `False`로, 이 경우 모델 스택이 생성되지 않습니다.
- `eval_metric`: 조기 중단 및 모델 비교에 사용할 지표입니다.
    - 이진 분류의 경우: `logloss`, `auc`, `f1`, `average_precision`, `accuracy`가 있으며 기본값은 'auto'로 설정하면 `logloss`가 사용됩니다.
    - 다중 클래스 분류의 경우: `logloss`, `f1`, `accuracy`가 있으며 기본값은 'auto'로 설정하면 `logloss`가 사용됩니다.
    - 회귀의 경우: `rmse`, `mse`, `mae`, `r2`, `mape`, `spearman`, `pearson`이 있으며 기본값은 'auto'로 설정하면 `rmse`가 사용됩니다.

- `validation_strategy`: 검증 유형에 대한 사전입니다. 현재 학습/테스트 분할 및 교차 검증이 지원됩니다.
    - 교차 검증 예시:
        ```python
        {
            "validation_type": "kfold",
            "k_folds": 5,
            "shuffle": True,
            "stratify": True,
            "random_seed": 123
        }
        ```
    - 학습/테스트 분할 예시:
        ```python
        {
            "validation_type": "split",
            "train_ratio": 0.75,
            "shuffle": True,
            "stratify": True
        }
        ```

- `explain_level`: 각 모델에 포함된 설명의 수준입니다.
    - `explain_level`이 0인 경우 설명이 생성되지 않습니다.
    - `explain_level`이 1인 경우 다음 설명이 생성됩니다: 중요도 그래프(순열 방법 사용), 결정 트리의 경우 트리 그래프 생성, 선형 모델의 경우 계수 저장.
    - `explain_level`이 2인 경우 다음 설명이 생성됩니다: 1과 동일하게 생성되며 추가로 SHAP 설명이 생성됩니다.

    'auto'로 설정하면 AutoML이 선택한 `mode`에 따라 설명을 생성합니다.
- `golden_features`: 골든 피처 사용 여부 및 추가할 골든 피처의 수입니다. 'auto', `True`/`False`, 또는 정수를 선택할 수 있습니다.
    - 'auto'로 설정하면 AutoML이 선택한 `mode`에 따라 골든 피처를 사용합니다:
        - `mode`가 "Explain"인 경우 `golden_features` = False.
        - `mode`가 "Perform"인 경우 `golden_features` = True.
        - `mode`가 "Compete"인 경우 `golden_features` = True.
    - `boolean` 값이 설정되면 골든 피처의 수가 자동으로 설정됩니다. min(100, max(10, 0.1 * 입력_피처_수))로 설정됩니다.
    - `int` 값이 설정되면 골든 피처의 수가 이 값으로 설정됩니다.

- `features_selection`: 피처 선택 여부입니다. 'auto' 또는 `True`/`False`를 선택할 수 있습니다.
    - 'auto'로 설정하면 AutoML이 선택한 `mode`에 따라 피처 선택을 수행합니다:
        - `mode`가 "Explain"인 경우 `features_selection` = False.
        - `mode`가 "Perform"인 경우 `features_selection` = True.
        - `mode`가 "Compete"인 경우 `features_selection` = True.

- `start_random_models`: 초기에 시도할 무작위 모델의 수입니다. 'auto' 또는 정수를 선택할 수 있습니다.
    - 'auto'로 설정하면 AutoML이 선택한 `mode`에 따라 무작위 모델의 수를 결정합니다:
        - `mode`가 "Explain"인 경우 `start_random_models` = 1.
        - `mode`가 "Perform"인 경우 `start_random_models` = 5.
        - `mode`가 "Compete"인 경우 `start_random_models` = 10.

- `hill_climbing_steps`: 힐 클라이밍 단계 수입니다. 'auto' 또는 정수를 선택할 수 있습니다.
    - 'auto'로 설정하면 AutoML이 선택한 `mode`에 따라 힐 클라이밍 단계 수를 결정합니다:
        - `mode`가 "Explain"인 경우 `hill_climbing_steps` = 0.
        - `mode`가 "Perform"인 경우 `hill_climbing_steps` = 2.
        - `mode`가 "Compete"인 경우 `hill_climbing_steps` = 2.

- `top_models_to_improve`: 힐 클라이밍 단계에서 개선할 상위 모델의 수입니다. 'auto' 또는 정수를 선택할 수 있습니다.
    - 'auto'로 설정하면 AutoML이 선택한 `mode`에 따라 상위 모델의 수를 결정합니다:
        - `mode`가 "Explain"인 경우 `top_models_to_improve` = 0.
        - `mode`가 "Perform"인 경우 `top_models_to_improve` = 2.
        - `mode`가 "Compete"인 경우 `top_models_to_improve` = 3.
- `boost_on_errors`: 이전 최상의 모델의 오류에 대한 부스트가 있는 모델을 학습할지 여부입니다. 기본적으로 `Compete` 모드에서 사용 가능합니다.
- `kmeans_features`: K-Means 생성 피처가 있는 모델을 학습할지 여부입니다. 기본적으로 `Compete` 모드에서 사용 가능합니다.
- `mix_encoding`: 혼합 인코딩이 있는 모델을 학습할지 여부입니다. 혼합 인코딩은 25개 이상의 범주를 가진 범주형에 대해 레이블 인코딩을 사용하고 다른 범주형에 대해 원-핫 이진 인코딩을 사용하는 인코딩입니다. 25보다 작은 기수를 가진 범주형 피처가 있을 경우에만 적용됩니다. 기본적으로 `Compete` 모드에서 사용 가능합니다.
- `max_single_prediction_time`: 단일 샘플 예측 시간의 제한입니다. 빠른 예측이 가능한 모델을 원할 경우 사용하세요. REST API로 사용되는 ML 파이프라인을 생성하는 데 이상적입니다. 시간은 초 단위입니다. 기본값(`max_single_prediction_time=None`)은 빠른 예측이 가능한 모델이 아닙니다. 단, `Perform` 모드의 경우 기본값은 0.5초입니다.
- `optuna_time_budget`: Optuna가 각 알고리즘을 조정하는 데 사용할 시간(초)입니다. 단일 알고리즘 조정 시간입니다. Xgboost와 CatBoost 두 알고리즘을 선택하고 optuna_time_budget=1000으로 설정하면 Xgboost는 1000초 동안 조정되고 CatBoost는 1000초 동안 조정됩니다. 또한 데이터 유형별로 조정이 수행됩니다(예: 원시 데이터 및 골든 피처가 삽입된 데이터). 이 매개변수는 `mode="Optuna"`일 때만 사용됩니다. `mode="Optuna"`로 설정하고 이 매개변수를 설정하지 않으면 3600초로 설정됩니다.

- `optuna_verbose`: `True`로 설정하면 Optuna 조정 세부 정보가 표시됩니다. 기본값은 `True`입니다.

`predict` 메서드와 `predict_all` 메서드의 차이점은 반환되는 예측 값의 형태입니다.

`predict` 메서드는 최상의 모델을 사용하여 예측을 수행하고 예측된 목표 값의 벡터를 반환합니다.

```python
predictions = automl.predict(X_test)
```

`predict_all` 메서드는 모든 학습된 모델을 사용하여 예측을 수행하고 각 모델에 대한 예측 결과를 포함하는 데이터프레임을 반환합니다.

```python
predictions = automl.predict_all(X_test)
```

반환된 데이터프레임에는 각 모델의 이름과 예측 값이 포함됩니다. 이를 사용하여 각 모델의 예측 결과를 비교하고 분석할 수 있습니다.

이것이 `predict` 메서드와 `predict_all` 메서드의 차이점입니다.
