## Quick Linear Regression!

```
#라이브러리 불러오기
import tensorflow as tf
from tensorflow import keras
import pandas as np
import numpy as np
```

```
x = np.array(range(20))
y = x * 2 - 1 

print(x)
print(y)
```

```
x.shape, y.shape #((20,) : 1차원 배열, (20, ) : 1차원 배열), 노드(원)와 layer가 있다
```

```
# 1번 세션 클리어 : 이미 만들어진 모델이 있다면 그 모델을 없애줘
keras.backend.clear_session()

# 2번 모델 선언(모델 발판 생성 : Sequential API방식으로 layer를 쌓는다)
model = keras.models.Sequential()

# 3번 모델 블록 조립(모델 레이어 쌓기)
model.add(keras.layers.Input(shape=(1,))) #x에 대한 layer
model.add(keras.layers.Dense(1)) #y에 대한 layer

# 4번 컴파일 
model.compile(loss = 'mse', #예측값과 실제값을 무엇으로 비교할까? 선형학습(지도학습) - 정답에 대한 길을 알려준다
              optimizer='adam') #무조건 adam을 이용, 얼마나 예쁜 방식으로 나갈지? 경사하향법을 이용해서 loss를 줄인다
```

```
model.fit( x, y, epochs=10, verbose=1) #몇번 학습시킬래? epochs, verbose -> 학습되는 과정을 보여준다 0, default 1, 2
```

```
y_pred = model.predict(x) #만들어낸 모델에서 예측값을 뽑아야한다
```

```
y_pred.reshape(-1)
```

```
y
```

```
model.predict(x).reshape(-1)
```

```
#라이브러리 불러오기
import tensorflow as tf
from tensorflow import keras
import numpy as np
import pandas as pd
```

```
x = np.array(range(20)) 
y = x * (-3) + 10

print(x)
print(y)
```

```
x.shape, y.shape #input과 output의 구조, 노드가 1개씩 있고 연결되어있다
```

```
#모델링 : Sequential API 순차적으로 쌓는다

#1. 세션 클리어 : 기존의 모델이 메모리에 남아있다면 지워줘!
keras.backend.clear_session()

#2. 모델을 쌓을 발판 생성
model = keras.models.Sequential()

#3. 모델 블록(layer) 조립
model.add(keras.layers.Input(shape = (1, )))
model.add(keras.layers.Dense(1))

#4. 컴파일
model.compile(loss = "mse", optimizer = "adam") #내 모델이 어떤 형태의 구조, 내 모델이 만든 예측값과 실제 정답을 무엇으로 비교할지,
#경사하강법으로 에러를 줄여나갈건데 그게 얼마나 예쁜 방식으로 내려가게 만들 것인지 결정하는 것
#에러를 MSE로 측정
```

```
model.fit(x, y, epochs = 10, verbose = 1) #1은 프로그래스bar가 기본, 2는 생략
```

```
y_pred = model.predict(x)
```

## Quick Logistic Regression! - 이진 분류

```
import tensorflow as tf
from tensorflow import keras

import numpy as np
```

```
x = np.array(range(20)) 
y = np.array([0]*10 + [1]*10)
print(x)
print(y)
```

```
x.shape, y.shape
```

```
#모델링 : Sequential API

#1. 세션 클리어 : 기존의 모델이 남아있다면 날려줘!
keras.backend.clear_session()

#2. 레이어 발판 생성 : model에 순차적으로 레이어를 쌓아가겠다는 의도!
model = keras.models.Sequential()

#3. 레이어 블록 조립 : model에 인풋 값을 받는 레이어를 넣음
model.add( keras.layers.Input(shape=(1,)) )

# model에 Dense 레이어를 넣을거야 (최초의 레이어) : weight를 곱하고, bias를 더해주는 과정
model.add( keras.layers.Dense(1, activation='sigmoid') )

#4. 컴파일 해주렴
model.compile(loss='binary_crossentropy', metrics=['accuracy'], #분류라서 정확도로 컴파일 해야한다.
              optimizer='adam') #기본적으로 경사하강법
```

```
# 데이터를 넣어서 학습시키자!
model.fit(x, y, epochs=10, verbose=1) #loss는 낮을수록 좋다

# 결과 출력해줘!
print(y)
print(model.predict(x)) #확률이 높을수록 1, 방문한다?
```
