```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
x = iris.data
y = iris.target
```

```python
x.shape, y.shape
```

```python
############
#train set / test set 분리
#조건1. test set size = 10%
#조건2. 난수 고정 = 2023
############

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.1, random_state = 2023)
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
#####################
# 라이브러리 불러오기
#####################
import tensorflow as tf
from tensorflow import keras
import numpy as np
```

```python
class_n = len(np.unique(y_train)) #y_train을 이용해서 클래스가 3개인 것을 확인했으니
```

```python
#######################
# 어떤 전처리 작업 필요 : One - Hot encoding
# Hint : y.shape
#y가 setosa, wesicolor, virginica 이렇게 가진다 각각으로 나눠서 이진분류한다 1개 노드-> 3개 노드
#softmax를 이용해서 3개 노드의 합을 1로 만들어준다
#######################
from tensorflow.keras.utils import to_categorical
y_train = to_categorical(y_train, class_n) #y_train을 이용해야 한다
y_test = to_categorical(y_test, class_n) #y_test을 이용해야 한다
```

```python
y_train.shape
```

```python
########
# 모델링
########
#1. 세션 클리어
keras.backend.clear_session()

#2. 모델 선언
model = keras.models.Sequential()

#3. 레이어 조립
model.add(keras.layers.Input(shape = (4,)))
model.add(keras.layers.Dense(3, activation = "softmax"))

#4. 컴파일
# model.compile(loss = "categorical_crossentropy", metrics = ["accuracy"], optimizer = "adam")
model.compile(loss = keras.losses.categorical_crossentropy, metrics = ["accuracy"], optimizer = "adam")
model.summary()
```

```python
###########
# 모델 학습
###########
model.fit(x_train, y_train, epochs = 10, verbose = 1)
```

```python
###########
# 모델 예측
###########
y_pred = model.predict(x)
```

```python
print(y_pred[:10])
print(y[:10])
```

# 히든 레이어 추가하여 모델링!
---
- 조건
    - 히든 레이어 2개를 추가하여 다시 모델링!
    - 히든 레이어 노드 수 : 32
    - 히든 레이어 activation = 'relu'

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
########
# 모델링
########
keras.backend.clear_session()

model = keras.models.Sequential()

model.add(keras.layers.Input(shape = (4,)))
model.add(keras.layers.Dense(32, activation = "relu"))
model.add(keras.layers.Dense(32, activation = "relu"))
model.add(keras.layers.Dense(3, activation = "softmax"))

model.compile(loss = "categorical_crossentropy", metrics = ["accuracy"], optimizer = "adam")

model.summary()
```

```python
###########
# 모델 학습
###########
model.fit(x_train, y_train, epochs = 10, verbose = 1)
```

```python
###########
# 모델 예측
###########
y_pred = model.predict(x_test)
```

```python
print(y_pred[:10])
print(y_test[:10])
```
