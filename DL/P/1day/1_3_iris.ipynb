```
from sklearn.datasets import load_iris
```

```
iris = load_iris()
```

```
x = iris.data
y = iris.target
```

```
print(x.shape, y.shape) #위 클래스는 3개인데 y는 1개이다.. 어떻게 접근해야 할까
print(iris.target_names)
```

```
# One-Hot Encoding
from tensorflow.keras.utils import to_categorical
```

```
y = to_categorical(y, 3) #한 번만 실행해야한다. 안 그럼 차원이 늘어난다
```

```
x.shape, y.shape
```

## 1번. Sequential API style로 모델링 하세요

- loss = 'categorical_crossentropy' 사용!

```
import tensorflow as tf
from tensorflow import keras
import numpy as np
import pandas as pd

#1. 이전 모델이 있는지 확인하고 지운다
keras.backend.clear_session()

#2. layer 생성
model = keras.models.Sequential()

#3. 레이어 블록 조립
model.add(keras.layers.Input(shape = (4, )))
model.add(keras.layers.Dense(3, activation = "softmax"))

#4. 컴파일
model.compile(loss = "categorical_crossentropy", metrics = ["accuracy"], optimizer = "adam")

#5. 요약
model.summary()
```

```
#선 하나하나가 전부 가중치
#Param 의미 : (4 + 1(편향값)) * 3 = 학습 가능한 parameter가 총 15개가 된다
#그 다음 레이어의 노드 수 = 첫 번째 레이어 : 4(노드) * 3(target) , 두 번째 레이어 : 3(target)
#연결주의 관점에서는 편향값을 일부러 준다
#또 다른 값을 줌으로써 모델의 학습을 일부러 흔든다. 그래서 다른 값도 더해서
#모델을 일반적으로 사용 가능하게 만드는 과정이라고 보면 된다
```

```
model.fit(x, y, epochs = 10, verbose = 1)
```

```
#Epoch 5/5 : mini-batch = 1 epoch에서 update 된 횟수
#10000개의 데이터 = Batch? 
```

```
y_pred = model.predict(x)
print(y_pred[:5]) #편향값이라고 기본적으로 내재된 값이라 생각
#setosa, versicolor, virginica
print(y[:5])
```
