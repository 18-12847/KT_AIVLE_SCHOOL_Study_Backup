```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer() #분류 문제임
```

```python
x = cancer.data
y = cancer.target
```

```python
x.shape, y.shape
```

```python
############
#train set / test set 분리
#조건 1. test set : 20%
#조건 2. 난수 고정 : 2023
############

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 2023)
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
#####################
# 라이브러리 불러오기
#####################
import tensorflow as tf
from tensorflow import keras
```

```python
########
# 모델링
########

#1번. 세션 클리어
keras.backend.clear_session()

#2번. 모델 선언
model = keras.models.Sequential()

#3번. 레이어 조립
model.add(keras.layers.Input(shape = (30, )))
model.add(keras.layers.Dense(1, activation = "sigmoid"))

#4번. 컴파일
model.compile(loss = keras.losses.binary_crossentropy, metrics = ["accuracy"], optimizer = "adam")

#양념
model.summary()
#sequential 특징은 Input layer가 summary에 출력 안된다
```

```python
###########
# 모델 학습
###########
model.fit(x_train, y_train, epochs = 10, verbose = 1)
#우리의 목적은 loss를 줄이는 것, accuracy는 보조 지표, 항상 서로 반비례 하지 않는다
```

```python
###########
# 모델 예측
###########

y_pred = model.predict(x_test)
print(y_pred[:10].reshape(-1))
print(y_test[:10])
```

# 히든 레이어 추가하여 모델링!
---
- 조건
    - 히든 레이어 2개를 추가하여 다시 모델링!
    - 히든 레이어 노드 수 : 32
    - 히든 레이어 activation = 'relu'

```python
x_train.shape, y_train.shape
```

```python
########
# 모델링 : Sequential API
########
#세션 클리어
keras.backend.clear_session()

#모델 선언
model = keras.models.Sequential()

#레이어 조립
model.add(keras.layers.Input(shape = (30, )))
model.add(keras.layers.Dense(32, activation = "relu"))
model.add(keras.layers.Dense(32, activation = "relu"))
model.add(keras.layers.Dense(1, activation = "sigmoid")) #분류 문제이기 때문!!!!

#컴파일
model.compile(loss = "binary_crossentropy", metrics = ["accuracy"], optimizer = "adam") #metrics : 보조 지표

#요약
model.summary()
```

```python
###########
# 모델 학습
###########

model.fit(x_train, y_train, epochs = 10, verbose = 1)
```

```python
###########
# 모델 예측
###########
y_pred = model.predict(x_test)
print(y_pred[:10].reshape(-1))
print(y_test[:10])
```
