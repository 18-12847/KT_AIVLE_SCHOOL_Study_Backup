# 1.환경 준비

- 기본 라이브러리와 대상 데이터를 가져와 이후 과정을 준비합니다.

```python
# 라이브러리 불러오기
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings(action='ignore')
%config InlineBackend.figure_format='retina'
```

```python
# 데이터 읽어오기
# path = 'https://raw.githubusercontent.com/Jangrae/csv/master/cars.csv'
path = 'https://bit.ly/CarsFile'
data = pd.read_csv(path)
```

# 2.데이터 이해

- 분석할 데이터를 충분히 이해할 수 있도록 다양한 탐색 과정을 수행합니다.

```python
# speed, dist 관계
plt.scatter(x='speed', y='dist', data=data)
plt.xlabel('Speed(mph)')
plt.ylabel('Dist(ft)')
plt.show()
```

# 3.데이터 준비

- 전처리 과정을 통해 머신러닝 알고리즘에 사용할 수 있는 형태의 데이터를 준비합니다.

**1) x, y 분리**

- target 변수를 명확히 지정합니다.
- target을 제외한 변수를 x 데이터프레임으로 선언합니다.
- y 데이터프레임은 target 변수만을 가집니다.

```python
# target 확인
target = 'dist'

# 데이터 분리
x = data.drop(target, axis=1)
y = data.loc[:, target]
```

**2) 학습용, 평가용 데이터 분리**

- 학습용, 평가용 데이터를 적절한 비율로 분리합니다.
- 반복 실행 시 동일한 결과를 얻기 위해 random_state 옵션을 지정합니다.

```python
# 모듈 불러오기
from sklearn.model_selection import train_test_split

# 7:3으로 분리
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)  
```

# 4.모델링

- 본격적으로 모델을 선언하고 학습하고 평가하는 과정을 진행합니다.

```python
# 1단계: 불러오기
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
# 2단계: 선언하기
model = LinearRegression() #성능을 높이려면 변수를 잘 정리해야 한다
```

```python
# 3단계: 학습하기
model.fit(x_train, y_train)
```

```python
# 4단계: 예측하기
y_pred = model.predict(x_test)
```

```python
# 5단계: 평가하기 - 평균 약 15피트 차이날 수 있다?
#평균에 비해 우리 모델이 약 55% 좋다는 뜻 - 그렇게 좋은 것은 아니다
print("MAE : ", mean_absolute_error(y_test, y_pred))
print("R2 : ", r2_score(y_test, y_pred))
```

- 회귀 계수를 살펴봅니다.

```python
# 회귀계수 확인
print(model.coef_, model.intercept_)
```

$$ \large Distance = 3.91 \times Speed - 16.37$$

# 5.기타

- 기타 필요한 내용이 있으면 진행합니다.

```python
#distance = a * speed + b
a = model.coef_ #회귀계수
b = model.intercept_ #편향
#x_train의 최솟값에서 최댓값 사이에 10개의 임의의 값을 만든다
speed = np.linspace(x_train.min(), x_train.max(), 10) 
dist = a * speed + b
```

```python
#학습 데이터 시각화
mean_dist = y_train.mean()
plt.figure(figsize = (4, 4))
plt.scatter(x_train, y_train)
plt.plot(speed, dist, color = "r")
plt.axhline(y = mean_dist, color = "r", linestyle = "--")
plt.show()
```

#이 선분이 학습데이터에서(x_train, y_train) 제일 오차가 적은 회귀 선분이다

```python
#평가 데이터 시각화
mean_dist = y_test.mean()
plt.figure(figsize = (4, 4))
plt.scatter(x_test, y_test) #y_test에서 y_pred로 바꾸면 회귀선 위에 찍힌다
plt.scatter(x_test, y_pred)
plt.plot(speed, dist, color = "r") #학습 데이터에서 최선의 선이다.
plt.axhline(mean_dist, color = "r", linestyle = "--")
plt.show()
```

```python
# 시각화, 정확하게 맞추지 못해서 어긋나는 부분이 있다
plt.plot(y_test.values, label='Actual')
plt.plot(y_pred, label='Predicted')
plt.legend()
plt.ylabel('Ozone')
plt.show()
```
